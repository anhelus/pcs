```python
import numpy as np
import matplotlib.pyplot as plt

from keras.preprocessing.text import Tokenizer
from keras.datasets import reuters
from keras import models, layers, losses
from keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data()
```

```python
# Data preprocessing
num_classes = np.amax(y_train) + 1          # Usato per garantire coerenza tra y_train ed y_test
tokenizer = Tokenizer(num_words=1000)
x_train = tokenizer.sequences_to_matrix(x_train, mode='binary')
x_test = tokenizer.sequences_to_matrix(x_test, mode='binary')
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
# overfitting ed underfitting con modello ad un layer ed 8 neuroni
model_small = models.Sequential()
model_small.add(
    layers.Dense(
        64,
        activation='relu',
        input_shape=(1000,)))
model_small.add(
    layers.Dense(
        num_classes,
        activation='softmax'))
model_small.compile(optimizer='adam',
            loss=losses.CategoricalCrossentropy(),
            metrics=['acc'])

history_small=model_small.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(x_test, y_test))
```

```python
# overfitting ed underfitting con modello a due layer ed 64 neuroni
model_medium = models.Sequential()
model_medium.add(
    layers.Dense(
        64,
        activation='relu',
        input_shape=(1000,)))
model_medium.add(
    layers.Dense(
        64,
        activation='relu'))
model_medium.add(
    layers.Dense(
        num_classes,
        activation='softmax'))
model_medium.compile(optimizer='adam',
            loss=losses.CategoricalCrossentropy(),
            metrics=['acc'])

history_medium=model_medium.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(x_test, y_test))
```

```python
# overfitting ed underfitting con modello ad un layer ed 8 neuroni
model_large = models.Sequential()
model_large.add(
    layers.Dense(
        64,
        activation='relu',
        input_shape=(1000,)))
model_large.add(
    layers.Dense(
        64,
        activation='relu'))
model_large.add(
    layers.Dense(
        64,
        activation='relu'))
model_large.add(
    layers.Dense(
        num_classes,
        activation='softmax'))
model_large.compile(optimizer='adam',
            loss=losses.CategoricalCrossentropy(),
            metrics=['acc'])

history_large=model_large.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(x_test, y_test))
```

```python
history_small_acc = history_small.history['acc']
history_small_val_acc = history_small.history['val_acc']
history_medium_acc = history_medium.history['acc']
history_medium_val_acc = history_medium.history['val_acc']
history_large_acc = history_large.history['acc']
history_large_val_acc = history_large.history['val_acc']

plt.plot(
    range(1, 51),
    history_small_acc,
    label='Accuracy (piccolo)')
plt.plot(
    range(1, 51),
    history_small_val_acc,
    label='Val accuracy (piccolo)')
plt.plot(
    range(1, 51),
    history_medium_acc,
    label='Accuracy (medio)')
plt.plot(
    range(1, 51),
    history_medium_val_acc,
    label='Val accuracy (medio)')
plt.plot(
    range(1, 51),
    history_large_acc,
    label='Accuracy (grande)')
plt.plot(
    range(1, 51),
    history_large_val_acc,
    label='Val accuracy (grande)')
plt.legend()
plt.grid()
plt.xlabel('Training epoch')
plt.ylabel('Accuracy (%)')
plt.show()
```

**Aggiunta di un parametro di regolarizzazione**

Proviamo ad aggiungere un parametro di regolarizzazione ai layer del modello.

```python
# overfitting ed underfitting con modello a due layer ed 64 neuroni
from keras import regularizers

model_medium_reg = models.Sequential()
model_medium_reg.add(
    layers.Dense(
        64,
        activation='relu',
        input_shape=(1000,),
        kernel_regularizer=regularizers.l2(0.001)))
model_medium_reg.add(
    layers.Dense(
        64,
        activation='relu',
        kernel_regularizer=regularizers.l2(0.001)))
model_medium_reg.add(
    layers.Dense(
        num_classes,
        activation='softmax'))
model_medium_reg.compile(optimizer='adam',
            loss=losses.CategoricalCrossentropy(),
            metrics=['acc'])

history_medium_reg=model_medium_reg.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(x_test, y_test))
```

```python
history_medium_reg_acc = history_medium_reg.history['acc']
history_medium_reg_val_acc = history_medium_reg.history['val_acc']

plt.plot(
    range(1, 51),
    history_small_acc,
    label='Accuracy (piccolo)')
plt.plot(
    range(1, 51),
    history_small_val_acc,
    label='Val accuracy (piccolo)')
plt.plot(
    range(1, 51),
    history_medium_acc,
    label='Accuracy (medio/reg)')
plt.plot(
    range(1, 51),
    history_medium_val_acc,
    label='Val accuracy (medio/reg)')
plt.plot(
    range(1, 51),
    history_medium_reg_acc,
    label='Accuracy (medio/reg)')
plt.plot(
    range(1, 51),
    history_medium_reg_val_acc,
    label='Val accuracy (medio/reg)')
plt.legend()
plt.grid()
plt.xlabel('Training epoch')
plt.ylabel('Accuracy (%)')
plt.show()
```

```python
model_medium_reg_drop = models.Sequential()
model_medium_reg_drop.add(
    layers.Dense(
        64,
        activation='relu',
        input_shape=(1000,)))
model_medium_reg_drop.add(layers.Dropout(0.8))
model_medium_reg_drop.add(
    layers.Dense(
        64,
        activation='relu',
        kernel_regularizer=regularizers.l2(0.001)))
model_medium_reg_drop.add(
    layers.Dense(
        num_classes,
        activation='softmax'))
model_medium_reg_drop.compile(optimizer='adam',
            loss=losses.CategoricalCrossentropy(),
            metrics=['acc'])

history_medium_reg_drop=model_medium_reg_drop.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(x_test, y_test))
```

```python
history_medium_reg_drop_acc = history_medium_reg_drop.history['acc']
history_medium_reg_drop_val_acc = history_medium_reg_drop.history['val_acc']

plt.plot(
    range(1, 51),
    history_small_acc,
    label='Accuracy (piccolo)')
plt.plot(
    range(1, 51),
    history_small_val_acc,
    label='Val accuracy (piccolo)')
plt.plot(
    range(1, 51),
    history_medium_acc,
    label='Accuracy (medio)')
plt.plot(
    range(1, 51),
    history_medium_val_acc,
    label='Val accuracy (medio)')
plt.plot(
    range(1, 51),
    history_medium_reg_drop_acc,
    label='Accuracy (medio/reg-drop)')
plt.plot(
    range(1, 51),
    history_medium_reg_drop_val_acc,
    label='Val accuracy (medio/reg-drop)')
plt.legend()
plt.grid()
plt.xlabel('Training epoch')
plt.ylabel('Accuracy (%)')
plt.show()
```

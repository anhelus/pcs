```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(42)
```

**Plot**

Definiamo la funzione `plot_x_y()` per generare un insieme casuale di coordinate $x$ ed $y$ che saranno unite a schermo mediante una spezzata.

```python
def plot_x_y():    
    x = np.arange(1, 6)
    y = rng.integers(low=0, high=10, size=5)
    fig, ax = plt.subplots()
    ax.plot(x, y)
    return fig, ax

fig, ax = plot_x_y()
```

Possiamo dare un nome agli assi ed alla figura, ed impostare una griglia di riferimento.

```python
fig, ax = plot_x_y()
ax.set_title('Il nostro primo plot')        # Imposto il titolo
ax.set_xlabel('Asse delle X')               # Imposto il label per l'asse X
ax.set_ylabel('Asse delle Y')               # Imposto il label per l'asse Y
ax.grid(True)                               # Imposto la griglia di riferimento
plt.show()
```

E' possibile impostare la griglia di riferimento su un unico asse specificando il parametro `axis`. Sottolineamo come valorizzandolo ad `'x'` sar generata soltanto la griglia per i valori di X (e quindi con un reticolato verticale e parallelo all'asse delle ordinate), mentre viceversa con la valorizzazione ad `'y'` sar generata la griglia per i valori di Y (e quindi con un reticolato orizzontale e parallelo all'asse delle ascisse).

```python
fig, ax = plot_x_y()
ax.grid(axis='x')                           # Imposto la griglia di riferimento sull'asse X
plt.show()
```

```python
fig, ax = plot_x_y()
ax.grid(axis='y')                           # Imposto la griglia di riferimento sull'asse X
plt.show()
```

**Esempio 1: Plot di pi funzioni**

Proviamo a plottare a schermo due funzioni, ovvero una retta ed una funzione sinusoidale.

```python
x = np.arange(0., 10., 0.01)
y_1 = 1 + 2 * x
y_2 = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y_1, label='Retta')
ax.plot(x, y_2, label='Funzione sinusoidale')

ax.set_title('Plot di due funzioni matematiche')
ax.set_xlabel('Asse x')
ax.set_ylabel('Asse y')

ax.legend()
ax.grid()

plt.show()
```

**Esempio 2: Subplot multipli**

Proviamo ad organizzare le funzioni che abbiamo plottato in precedenza in due diversi subplot.

```python
fig, (ax_1, ax_2) = plt.subplots(2, 1)
fig.suptitle('Due subplot di pi funzioni matematiche')
# Primo subplot
ax_1.plot(x, y_1, label='Retta')
ax_1.set_ylabel('Asse y')
ax_1.legend()
ax_1.grid()
# Secondo subplot
ax_2.plot(x, y_2, label='Funzione sinusoidale')
ax_2.set_xlabel('Asse x')
ax_2.set_ylabel('Asse y')
ax_2.legend()
ax_2.grid()
# Mostro la figura
plt.show()
```

**Esempio 3: Istogramma**

Proviamo a visualizzare a schermo un istogramma.

```python
x = rng.integers(low=0, high=100, size=1000)
fig, ax = plt.subplots()
ax.hist(x, edgecolor='black', linewidth=1.2, density=True)

ax.set_xlabel('Bin')
ax.set_ylabel('Conteggio dei singoli elementi')
ax.set_title('Esempio di istogramma')

plt.show()
```

**Esempio 4: Scatter plot**

Usiamo il codice seguente per visualizzare a schermo uno scatter plot.

```python
rng = np.random.default_rng(42)
x = np.arange(1, 6)
y = rng.integers(low=0, high=10, size=5)
fig, ax = plt.subplots()
ax.set_xlabel('Asse x')
ax.set_ylabel('Asse y')
ax.set_title('Esempio di scatter plot')
ax.scatter(x, y)
plt.savefig('scatter.png')
plt.show()
```

**La funzione `close()`**

Vediamo i risultati di un plot inserito in un ciclo utilizzando o meno la funzione `close()` ad ogni iterazione.

```python
for i in range(5):
    plt.plot(x, y*i)
plt.savefig('not_close.png', dpi=600)
plt.show()
```

```python
for i in range(5):
    plt.plot(x, y*i)
    if i == 4:
        plt.savefig('closed.png')
    plt.show()
    plt.close()
```

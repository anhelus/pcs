```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm, uniform


def plot_distribution(dtype):
    x_1 = np.linspace(dtype.ppf(0.01), dtype.ppf(0.99), 100)
    r_1 = dtype.rvs(size=1000)
    plt.plot(x_1, dtype.pdf(x_1))
    plt.hist(r_1, density=True)
    plt.show()

# x_1 = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
# r_1 = norm.rvs(size=1000)

# plt.plot(x_1, norm.pdf(x_1))
# plt.hist(r_1, density=True)
# plt.show()
plot_distribution(norm)
```

```python
plot_distribution(uniform)
```

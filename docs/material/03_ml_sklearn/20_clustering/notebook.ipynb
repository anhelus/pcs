```python
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
# Caricamento dei dati e fitting
data = load_iris(as_frame=True)
kmeans = KMeans(n_clusters=3)
kmeans.fit(data.data.values)
sns.relplot(
    data=data.data,
    x='sepal length (cm)',
    y='sepal width (cm)',
    hue=data.target)
```

```python
y_pred = kmeans.predict(data.data.values)
sns.relplot(
    data=data.data,
    x='sepal length (cm)',
    y='sepal width (cm)',
    hue=y_pred)
```

```python
from sklearn.metrics import adjusted_rand_score

adjusted_rand_score(y_pred, data.target)
```
